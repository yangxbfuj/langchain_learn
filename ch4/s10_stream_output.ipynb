```python
from typing import TypedDict


class State(TypedDict):
    topic: str
    story: str


def refine_topic(state: State) -> State:
    return {"topic": state["topic"] + "的故事"}


def generate_story(state: State) -> State:
    return {"story": f"这是一个关于{state['topic']}的故事"}


from langgraph.graph import END, START, StateGraph

graph = (
    StateGraph(State)
    .add_node("refine_topic", refine_topic)
    .add_node("generate_story", generate_story)
    .add_edge(START, "refine_topic")
    .add_edge("refine_topic", "generate_story")
    .add_edge("generate_story", END)
    .compile()
)

print("--")
for chunk in graph.stream({"topic": "通话"}, stream_mode="updates"):
    print(chunk)

print("--")
async for chunk in graph.astream({"topic": "通话"}, stream_mode="updates"):
    print(chunk)

print("--")
async for chunk in graph.astream({"topic": "通话"}, stream_mode="debug"):
    print(chunk)

print("--")
async for chunk in graph.astream({"topic": "通话"}, stream_mode="values"):
    print(chunk)

print("--")
async for chunk in graph.astream({"topic": "通话"}, stream_mode="messages"):
    print(chunk)
    
```
