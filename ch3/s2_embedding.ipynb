```python
import os

document_text = """
锄禾日当午，
汗滴禾下土。
谁知盘中餐，
粒粒皆辛苦。
"""


from langchain_community.embeddings import DashScopeEmbeddings

embeddings = DashScopeEmbeddings(model="text-embedding-v4", dashscope_api_key=os.getenv("QWEN_API"))

doc_embedding = embeddings.embed_documents(document_text)

print(f"文档：向量维度为: {len(doc_embedding)}")
print(f"文档：前10个向量值为: {doc_embedding[:10]}")

query = "锄禾什么？"
query_embedding = embeddings.embed_query(query)

print(f"查询：向量维度为: {len(doc_embedding)}")
print(f"查询：前10个向量值为: {doc_embedding[:10]}")
```

```python
import os

from langchain_community.document_loaders import PyPDFLoader
from langchain_core.vectorstores import InMemoryVectorStore

file_path = f"{os.getcwd()}/files/RD08雷达显控对外通讯协议V2.11（公开版）.pdf"
loader = PyPDFLoader(file_path)
docs = loader.load()

embeddings = DashScopeEmbeddings(model="text-embedding-v4", dashscope_api_key=os.getenv("QWEN_API"))
vector_store = InMemoryVectorStore(embeddings)

# 添加
ids = vector_store.add_documents(docs)
query = "如何控制雷达水平角？"
from typing import List

from langchain_core.documents import Document

result: List[Document] = vector_store.similarity_search(query=query, k=2)

print(f"------- 查询： {query} ——————————————")
for i, d in enumerate(result, 1):
    print(f"{i} {d.page_content}")
```
