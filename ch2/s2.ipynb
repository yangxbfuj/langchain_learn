```python
from langchain_core.prompts import ChatPromptTemplate

chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的AI助理"),
    ("human", "请用简短的语言回答我: {question}")
])

pv = chat_prompt.format_prompt(question="什么是日食？")
print("===  渲染结果 ===")
for msg in pv.to_messages():
    print(msg.content)
print("===  格式化结果 ===")
print(pv.to_string())
```

```python
from langchain.chat_models import init_chat_model
from langchain.messages import SystemMessage, HumanMessage, AIMessage
import os
model = init_chat_model(
    model="deepseek-chat",
    temperature=0.7,
    api_key=os.environ.get("DEEPSEEK_API"),
    base_url="https://api.deepseek.com/v1",
    model_provider="openai"
)
print("===  第一轮对话 ===")
msg_round_1 = [
    SystemMessage(content="你是一个中国古典诗人， 擅长写七言绝句"),
    HumanMessage(content="请写一首关于AI的七言绝句"),
]
response_round_1 = model.invoke(msg_round_1)
print(f"AI回复: {response_round_1.content}")

print("===  第二轮对话 ===")
msg_round_2 = [*msg_round_1, response_round_1, HumanMessage(content="请解释这首诗的含义")]
response_round_2 = model.invoke(msg_round_2)
print(f"AI回复: {response_round_2.content}")
```

```python
from langchain.messages import AIMessage, ToolMessage
# ---1. 模拟工具调用请求 ---
model_tool_call = AIMessage(
    content="",
    tool_calls=[
        {"id": "call_123", "name": "calculator", "args": {"a": 1, "b": 2}}
    ]
)

# ---2. 模拟外部工具执行并返回结果
calculator_result = "5"

# ---3. 创建 ToolMessage, 并将结果反馈给模型
tool_feedback_msg = ToolMessage(
    content=calculator_result,
    tool_call_id="call_123",
)

print("===  格式打印 ===")
print(f"类型: {tool_feedback_msg.type}")
print(f"内容: {tool_feedback_msg.content}")
print(f"工具调用ID: {tool_feedback_msg.tool_call_id}")
```
