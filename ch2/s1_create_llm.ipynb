```python
import dotenv

dotenv.load_dotenv()
```

```python
# 调用 DeepSeek 模型
import os

from langchain.chat_models import init_chat_model

model = init_chat_model(
    "deepseek-chat",
    temperature=0.7,
    api_key=os.environ.get("DEEPSEEK_API"),
    base_url="https://api.deepseek.com/v1",
    model_provider="openai",
)

for chunk in model.stream("你是什么模型?你能识别哪些数据?生成哪些数据?"):
    print(chunk.content, end="", flush=True)
```

```python
# 调用 GLM 模型
import os

import dotenv
from langchain.chat_models import init_chat_model

dotenv.load_dotenv()
model = init_chat_model(
    model="glm-5.2",
    temperature=0.7,
    api_key=os.environ.get("ZP_API"),
    base_url="https://open.bigmodel.cn/api/paas/v4/",
    model_provider="openai",
)

for chunk in model.stream("你是什么模型?你能识别哪些数据?生成哪些数据?你能生成图片吗？"):
    print(chunk.content, end="", flush=True)
```

```python
# 调用 QWEN 模型
import os

import dotenv
from langchain_community.llms.tongyi import Tongyi

dotenv.load_dotenv()
model = Tongyi(
    model="qwen-plus",
    temperature=0.7,
    api_key=os.environ.get("QWEN_API"),
)

for chunk in model.stream("你是什么模型?你能识别哪些数据?生成哪些数据?你能生成图片吗？"):
    print(chunk, end="", flush=True)
```

```python
# 调用 QWEN 模型
import os

import dotenv
from dashscope import ImageSynthesis
from langchain_community.llms.tongyi import Tongyi

dotenv.load_dotenv()
result = ImageSynthesis.call(
    model="wanx-v1",
    size="1024*1024",
    prompt="一只可爱的猫咪在草地上玩耍, 图片风格需要写实, 画面清晰, 色彩鲜艳",
    api_key=os.environ.get("QWEN_API"),
)
print(result.status_code)
if result.status_code == 200:
    output = result.output
    if "url" in output["results"][0]:
        from IPython.display import Image, display

        display(Image(url=output["results"][0]["url"]))
```
