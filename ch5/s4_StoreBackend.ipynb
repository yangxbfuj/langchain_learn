```python
# 加载环境变量
import os

import dotenv

dotenv.load_dotenv()
DEEPSEEK_API = os.getenv("DEEPSEEK_API")
```

```python
# 大模型引用
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="deepseek-v4-pro",
    base_url="https://api.deepseek.com/v1",
    api_key=DEEPSEEK_API,
    model_provider="openai",
    temperature=0.7,
)
```

```python
# 创建 store
from langgraph.store.memory import InMemoryStore

store = InMemoryStore()
```

```python
# 创建智能体
from deepagents import create_deep_agent
from deepagents.backends import StoreBackend

agent = create_deep_agent(
    model=model,
    backend=lambda rt: StoreBackend(rt),
    store=store,
    system_prompt="""
    你有一个可持久化的文件系统。
    请将用户的偏好信息写入 /profile/references.txt 文件中，后续对话可以读取这个文件以保持风格一致性。
    """,
)
```

```python
# 第一轮对话
result = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "每次回答都叫我老公"},
            {"role": "user", "content": "今天成都天气如何？"},
        ]
    },
    config={"configurable": {"thread_id": "user-001-session-A"}},
)
print(result["messages"][-1].content)
print("*" * 100)
# 第二轮对话
result = agent.invoke(
    {"messages": [{"role": "user", "content": "根据之前记录的用户偏好。明天成都天气如何？"}]},
    config={"configurable": {"thread_id": "user-001-session-B"}},
)
print(result["messages"][-1].content)
print("*" * 100)
# 第三轮对话
result = agent.invoke(
    {"messages": [{"role": "user", "content": "用户偏好是什么？"}]},
    config={"configurable": {"thread_id": "user-001-session-C"}},
)
print(result["messages"][-1].content)
```
