```python
# 加载环境变量

import dotenv

dotenv.load_dotenv()
```

```python
# 创建大模型引用

import os

from langchain.chat_models import init_chat_model

llm = init_chat_model(
    model="deepseek-v4-pro",
    base_url="https://api.deepseek.com/v1",
    api_key=os.getenv("DEEPSEEK_API"),
    model_provider="openai",
    temperature=0.7,
)
```

```python
def print_file(content: str):
    """打印文件的内容，当用户需要打印某个文件内容时使用

    Args:
        content (str): 文件内容，需要utf-8编码
    """
    print("-" * 100)
    print(content)
    print("-" * 100)
```

```python
# 创建智能体

from deepagents import create_deep_agent

agent = create_deep_agent(
    model=llm,
    tools=[print_file],
    system_prompt="""
    你是一个善于规划的 Deep Agent.
    请把你对当前任务的执行计划写入 /scratch/plan.md文件中，然后基于该文件给我一个简单的说明。
    """,
)

config = {"configurable": {"thread_id": "state-demo-001"}}
```

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "为一个 Deep Agent 入门教程列一个章节计划"}]}, config=config
)
print(result["messages"])
result = agent.invoke(
    {"messages": [{"role": "user", "content": "使用工具打印 /scratch/plan.md 的内容"}]}, config=config
)
print(result["messages"])
result = agent.invoke(
    {"messages": [{"role": "user", "content": "基于你之前写入的 /scratch/plan.md, 精简内容"}]}, config=config
)
print(result["messages"])
```
