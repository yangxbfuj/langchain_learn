```python
import os

import dotenv
from langchain.chat_models import init_chat_model

# 加载坏境变量
dotenv.load_dotenv()

# 创建大模型引用
llm = init_chat_model(
    model="deepseek-v4-pro",
    api_key=os.getenv("DEEPSEEK_API"),
    base_url="https://api.deepseek.com/v1",
    model_provider="openai",
    temperature=0.7,
)


from tavily import TavilyClient

# tavily 客户端
tavily_client = TavilyClient(api_key=os.getenv("TAVILY_KEY"))

from typing import Literal


# tavily 搜索，为 Agent 提供工具
def tavily_search(
    query: str, max_results: int = 3, topic: Literal["general", "news"] = "general", include_raw_content: bool = False
):
    """互联网检索工具， 允许智能体在互联网上收集信息

    Args:
        query (str): 查询字符串
        max_results (int, optional): 查询最大数量. Defaults to 3.
        topic (Literal[&quot;general&quot;, &quot;news&quot;], optional): 检索主题类型. Defaults to "general".
            - "general": 通用检索
            - "news": 新闻检索
        include_raw_content (bool, optional): 是否包含原始内容. Defaults to False.

    Returns:
        Tavily 检索结果字典， 包含检索结果列表和相关元数据
    """
    print(f"检索内容: {query}")
    return tavily_client.search(
        query=query, max_results=max_results, include_raw_content=include_raw_content, topic=topic
    )


# 系统提示词
research_instructions = """
角色: 你是一个负责研究任务的 Deep Agent

在处理复杂任务时， 你需要:
1. 用待办工具规划要执行的步骤;
2. 按照步骤内容调用工具（如 tavily_search ) 收集信息;
3. 必要时将中间结果写入虚拟文件系统以便反复查看;
4. 输出结构化、清晰、可执行的中文结论。
"""

from deepagents import create_deep_agent

# 创建智能体
agent = create_deep_agent(
    model=llm,  # 调用的大模型
    tools=[tavily_search],  # 给他提供的工具
    system_prompt=research_instructions,  # 系统提示词
)

# 回答用户问题
async for msg in agent.astream({"messages": [{"role": "user", "content": "预测2026世界杯冠军队伍"}]}):
    print(msg)
```

```python
result = agent.invoke({"messages": [{"role": "user", "content": "预测2026世界杯冠军队伍"}]})
```
